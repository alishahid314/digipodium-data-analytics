```python
from random import random , randint,choice,shuffle
```

```python
print('get a random value between 0 and 1:',random())
print('get a random value between 0 and 1:',randint(50,100))
print('select a random item',choice(['jhon','jack','jill','jamie']))
names=['jhon','jack','jill','jamie']
shuffle(names)
print('shuffle a list',names)
```
