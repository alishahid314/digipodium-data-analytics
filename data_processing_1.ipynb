```python
def price_clean_column(price):
    if isinstance(price,str): #check if price variable is string
        price=price.replace(',','') #remove comma
        price=price.strip() #remove white space
        if price.isnumeric():
            price=int(price)
            return price
        if len(price)==0:
            return np.nan

    return price  
```

```python
df['price'].apply(price_clean_column)
```

```python
df['cleaned_price'] = df['price'].apply(price_clean_column)
```

```python
df.sort_values(by='title',ascending=False,inplace=True)
df
```

```python
df.sort_values(by='cleaned_price',ascending=True,inplace=False)
df
```
