```python
import pandas as pd
import numpy as np
import seaborn as sns #seaborn is used statistical graph
import matplotlib.pyplot as plt 

```

```python
iris=sns.load_dataset('iris')
iris.head() #iris is flower
```

```python
sns.histplot(iris)
```

#single column analysis

```python
sns.histplot(iris,x='sepal_length')
```

```python
sns.histplot(iris,x='sepal_length',bins=20,kde=True)
```

```python
sns.histplot(iris,x='sepal_width',bins=20,kde=True) #kde=kernal density
```

```python
#to resize
fig,ax=plt.subplots(figsize=(5,3))
```

```python
fig,ax=plt.subplots(figsize=(5,3))
sns.histplot(iris,x='sepal_width',bins=20,kde=True)
plt.savefig('sepal_width_hist.png',dpi=300)
```

```python
#for petal length
sns.histplot(iris,x='petal_width')
```

```python
fig,ax=plt.subplots(figsize=(5,3))
sns.histplot(iris,x='petal_width',bins=20,kde=True)
plt.savefig('petal_width_hist.png',dpi=300)
```

```python
#to collect all graph  in sigle graph/multiple 
fig,ax= plt.subplots(figsize=(20,5))
```

```python
#to create 4 graph

fig,ax= plt.subplots(figsize=(20,5),ncols=4)
```

```python
#to create multiple graph data in single graph
sns.set_palette('rainbow')
fig,ax= plt.subplots(figsize=(20,5),ncols=4)
sns.histplot(iris,x='sepal_width',bins=15,kde=True,ax=ax[0],color='red')
sns.histplot(iris,x='petal_width',bins=15,kde=True,ax=ax[1])
sns.histplot(iris,x='sepal_length',bins=15,kde=True,ax=ax[2])
sns.histplot(iris,x='petal_length',bins=15,kde=True,ax=ax[3])
```

```python
sns.set_palette('rainbow')
fig,ax= plt.subplots(figsize=(15,8),ncols=2,nrows=2)
sns.histplot(iris,x='sepal_width',bins=15,kde=True,ax=ax[0,0],color='red')
sns.histplot(iris,x='petal_width',bins=15,kde=True,ax=ax[0,1])
sns.histplot(iris,x='sepal_length',bins=15,kde=True,ax=ax[1,0])
sns.histplot(iris,x='petal_length',bins=15,kde=True,ax=ax[1,1])
```
